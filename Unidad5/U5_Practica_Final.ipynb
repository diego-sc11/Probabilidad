##Regresion lineal y correlación

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

##Grupo 1


```python
df1=pd.read_csv("grupo1.csv")
print(df1.head())
print(df1.info())
print(df1.describe())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df1['Calificacion'], bins=10)
plt.title('Histograma de Calificaciones Grupo 1')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# MODELO DE REGRESION LINEAL
x = df1[['HorasEstudio']]
y = df1['Calificacion']
modelo = LinearRegression()
modelo.fit(x, y)
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Grupo 1:")
print(f'Pendiente: {pendiente:.4f}')
print(f'Intercepto: {intercepto:.4f}')

print("Ecuación de regresión:")
print(f"y = {pendiente:.4f}x + {intercepto:.4f}")

```

```python
# CORRELACION DE PEARSON
correlacion, p_value = pearsonr(df1['HorasEstudio'],
                                    df1['Calificacion'])
print("Coeficiente de correlación de Pearson:")
print(f"r = {correlacion}")

print("Valor p:")
print(f"p = {p_value}")
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df1['HorasEstudio'],
                df1['Calificacion'],
                label='Datos')

y_pred = modelo.predict(x)

plt.plot(df1['HorasEstudio'],
             y_pred,
             linewidth=2,
             label='Línea de regresión')

plt.title('Regresión Lineal grupo 1')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```
