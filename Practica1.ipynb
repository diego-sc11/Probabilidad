```python
from statistics import *
```

```python
# Lista de calificaciones del grupo a de probabilidad y estadistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
# la funcion type nos devuelve el tipo de dato de la variable
type(grupoa)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
promedio2 = mean(grupoa)
print(promedio2)
```

```python
# Determinar el valor de la mediana con la funcion median

mediana = median(grupoa)
print(mediana)
```

```python
# Determinar el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

```python
# En caso de que exista mas de una moda, se puede emplear la funcion multimode para sacar multiples modas que puedan existir

moda2 = multimode(grupoa)
print(moda2)
```

```python
# Definimos una variable llamada tiempos100mts de tipo lista, que contiene los tiempos que tardaron los 10 alumnos para correr los 100 metros
tiempos100mts = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.10, 17.20]
print(tiempos100mts)
```

```python
#Tiempo promedio, mediana y moda de los corredores
print("El tiempo promedio de los corredores es=> ", mean(tiempos100mts))
print("El tiempo mediano de los corredores es=> ", median(tiempos100mts))
print("El tiempo moda de los corredores es=> ", mode(tiempos100mts))
```
