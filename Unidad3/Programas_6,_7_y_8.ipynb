#Programa 6

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

#Verifique que las probabilidades sumen 1
print("Suma de probabilidades:", sum(p))

# Graficar distribución
plt.stem(x, p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Celulas anormales")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()


#Calcular valor esperado
valor_esperado = 0
for i in range(len(x)):
  xi = x[i]     # valor de la variable
  pi = p[i]     #probabilidad de ese valor
  producto = xi * pi            # Multiplicar valor por su probabilidad
  valor_esperado += producto      # sumar al acumulador
  # Mostrar cada paso
  print(f"x = {xi}, p = {pi}, x*p = {producto:.2f}")
#Mostrar resultado final
print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

```python
import math

# Calcular E(X)
esperanza = 0
for i in range(len(x)):
    esperanza += x[i] * p[i]

# Calcular E(X^2)
esperanza_cuadrado = 0
for i in range(len(x)):
  esperanza_cuadrado += (x[i]**2) * p[i]

# Calcular varianza y desviacion estandar
varianza = esperanza_cuadrado - esperanza**2
desviacion = math.sqrt(varianza)
# Mostrar resultados
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviacion estandar = {desviacion:.2f}")
```

```python
# Funcion de distribucion acumulada
import numpy as np
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4, 5]    #Valores posibles de la variable
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]    #Probabiliades correspondientes
# Calcular funcion acumulada
F = np.cumsum(p)

# Mostrar resultados
for xi, fi in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")

# Graficar
plt.step(x, F, where='post')
plt.title("Funcion de distribucion Acumulada (FDA)")
plt.xlabel("Celulas anormales")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
