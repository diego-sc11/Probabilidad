```python
#Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos
#y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen
#normalmente. Hallar lo siguiente:
#La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
#Cuantos estudiantes pesan más de 68 y menos de 72 kg.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#Generar datos para la distribucion normal

media = 70  #Media
desv_std = 4  # Desviacion estandar
num_datos = 500

# generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
plt.figure(figsize=(6, 4))

# Parametros de la distribucion normal
media = 70  #Media
desv_std = 4 # Desviacion estandar
peso_limite_inferior = 68 # Peso limite
peso_limite_superior = 72 # Peso limite
total_estudiantes = 500

prob_entre_68_72 = norm.cdf(peso_limite_superior, media, desv_std) - norm.cdf(peso_limite_inferior, media, desv_std)

#Cuantos estudiantes pesan más de 68 y menos de 72 kg.
estudiantes_entre_68_72 = total_estudiantes * prob_entre_68_72

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_entre_68_72)
print("Numero de estudiantes que pesan entre 68 y 72 kg:", estudiantes_entre_68_72)

# Calcular la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribucion normal
plt.plot(x, y, 'b', linewidth = 2)
plt.fill_between(x, y, where=(x >= peso_limite_inferior) & (x <= peso_limite_superior), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68 y menos de 72. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
