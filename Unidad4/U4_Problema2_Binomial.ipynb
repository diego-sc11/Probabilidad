##Realizar un programa en python que resuelva el siguiente problema:
* Un fabricante, de podadoras para el pasto compra motores de 1 hp y 2 ciclos, en lotes de 1,000, a un proveedor.
* El fabricante equipa cada una de las podadoras producidas por su planta con uno de los motores.
* La historia muestra que la probabilidad de que cualquier motor del proveedor resulte no satisfactorio es .001.
* **En un embarque de 1,000 motores, Calcular**
* ¿cuál es la probabilidad de que, ninguno sea defectuoso? ,
* ¿cuál es lla probabilidad de que, tres sean defectuosos
* ¿cuál es la probabilidad de que, cuatro sean defectuosos?

##Es de tipo binomial, ya que cuenta exitos en un numero fijo de ensayos

```python
from scipy.stats import binom

# Parámetros
num_motores = 1000
prob_defectuoso = 0.001

# a) Ninguno defectuoso
prob_0 = binom.pmf(0, num_motores, prob_defectuoso)

# b) Exactamente 3 defectuosos
prob_3 = binom.pmf(3, num_motores, prob_defectuoso)

# c) Exactamente 4 defectuosos
prob_4 = binom.pmf(4, num_motores, prob_defectuoso)

print("Probabilidad de 0 motores defectuosos =", prob_0)
print("Probabilidad de 3 motores defectuosos =", prob_3)
print("Probabilidad de 4 motoresdefectuosos =", prob_4)
```
