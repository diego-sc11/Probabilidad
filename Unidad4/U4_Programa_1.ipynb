```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#Generar datos para la distribucion normal

media = 70  #Media
desv_std = 4  # Desviacion estandar
num_datos = 500

# generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
# # Crear un rango de valores para crear una densidad
x = np.linspace(media -3 * desv_std, media +3 * desv_std, num_datos)

    #Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi))*np.exp(-(x - media)**2 / (2 * desv_std **2))

# Para no imprimir toda la lista de valores (500), desplegamos los primeros 50

print("Imprimir los primeros 50 valores de x", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
    # Visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label= 'Normal: (media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de densidad de probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2, )

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

#Parametros de la distribucion normal
media = 70 #Media
desv_std = 4 #Desviacion estandar
total_estudiantes = 500

#Variable para calculo de la probabilidad de que un estudiante que pese menos de 75
peso_limite = 75

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 75 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
# Calculo del numero de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg =  total_estudiantes *prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

# Graficar la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth = 2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de qu e un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
#Cuantos estudiantes pesan menos de 62 kg
plt.figure(figsize=(6, 4))

# Parametros de la distribucion normal
media = 70  #Media
desv_std = 4 # Desviacion estandar
total_estudiantes = 500

#VARIABLE para calculo de la probabilidad de que un estudiante pese menos de 62 kg
peso_limite = 62

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 62 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)
# Calculo del numero de estudiantes que pesan menos de 62 kg
estudiantes_menos_62kg =  total_estudiantes *prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Numero de estudiantes que pesan menos de 62 kg:", estudiantes_menos_62kg)

# Calcular la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribucion normal
plt.plot(x, y, 'b', linewidth = 2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
#Cuantos estudiantes pesan mas de 79 kg
plt.figure(figsize=(6, 4))

# Parametros de la distribucion normal
media = 70  #Media
desv_std = 4 # Desviacion estandar
peso_limite = 79 # Peso limite
total_estudiantes = 500

# Calculamos la probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese mas de 79 kg es complementaria
prob_mas_79kg =  1 - prob_menos_79kg

# Calculamos el numero aproximado de estudiantes que pesan mas de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilidad de que un estudiante pese mas de 79 kg:", prob_mas_79kg)
print("Numero de estudiantes que pesan mas de 79 kg:", estudiantes_mas_79kg)

# Calcular la funcion de densisdad de probabilidad de la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la distribucion normal
plt.plot(x, y, 'b', linewidth = 2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 79 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
