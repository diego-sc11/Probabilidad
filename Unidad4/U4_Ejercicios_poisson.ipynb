##Distribucion de poisson
##Ejercicio 1

```python
from scipy.stats import

#Parametros de la distribucion de poisson
media_accidentes = 2

#Parte 1: Probabilidad de que no haya accidentes en una semana
probabilidad_0_accidentes = poisson.pmf(0, media_accidentes)
print("1. Probabilidad de que no haya accidentes en 1 semana:", probabilidad_0_accidentes)

#Parte 2: Probabilidad de que haya tres accidentes en 2 semanas
#Convertimos la media a una media para 2 semanas
media_2_semanas = media_accidentes * 2

#Calculamos la probabilidad acumulada hasta 3 accidentes en 2 semanas
probabilidad_3_o_menos_accidentes = poisson.cdf(3, media_2_semanas)

print("2. Probabilidad de que a lo sumo haya tres accidentes en 2 semanas:", probabilidad_3_o_menos_accidentes)
```

##Ejercicio 2

* Una compañia de seguros de vida asegura las vidas de 5000 hombres de 42 años de edad.
* Si estudios aciales muestran que la probabilidad de que cualquier hombre de 42 años muera en un año determinado es .001 encuentre la probabilidad exacta de que la compañia tendra que pagar x = 4 reclamaciones durante un año determinado

```python
from scipy.stats import poisson

#Parametros de la distribucion de Poisson
media_muertes = 5000 * 0.001

# Numero de reclamaciones
x = 4

# Calcular la probabilidad de que la compañia tenga que pagar 4 reclamaciones en un año determinado
probabilidad_4_reclamaciones = poisson.pmf(x, media_muertes)
print("La probabilidad de que la compañia tenga que pagar 4 reclamaciones es De:", probabilidad_4_reclamaciones)
```
