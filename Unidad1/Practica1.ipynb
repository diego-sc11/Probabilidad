```python
from statistics import *
```

```python
# Lista de calificaciones del grupo a de probabilidad y estadistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
# la funcion type nos devuelve el tipo de dato de la variable
type(grupoa)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
promedio2 = mean(grupoa)
print(promedio2)
```

```python
# Determinar el valor de la mediana con la funcion median

mediana = median(grupoa)
print(mediana)
```

```python
# Determinar el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

```python
# En caso de que exista mas de una moda, se puede emplear la funcion multimode para sacar multiples modas que puedan existir

moda2 = multimode(grupoa)
print(moda2)
```

```python
# Definimos una variable llamada tiempos100mts de tipo lista, que contiene los tiempos que tardaron los 10 alumnos para correr los 100 metros
tiempos100mts = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.10, 17.20]
print(tiempos100mts)
```

```python
#Tiempo promedio, mediana y moda de los corredores
print("El tiempo promedio de los corredores es=> ", mean(tiempos100mts))
print("El tiempo mediano de los corredores es=> ", median(tiempos100mts))
print("El tiempo moda de los corredores es=> ", mode(tiempos100mts))
```

```python
#importar numeros aleatoreos
import random
fabrica1 = []
for i in range(100):
  numero_aleatoreo = round( random.uniform(50, 100),2)
  fabrica1.append(numero_aleatoreo)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatoreo = round( random.uniform(70, 100),2)
  fabrica2.append(numero_aleatoreo)
print(fabrica2)
```

```python
# libreria de python para graficar valores
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
import matplotlib.pyplot as plt
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("La media de la fabrica 1 es: ", media1)
print("La mediana de la fabrica 1 es: ", mediana1)
print("La moda de la fabrica 1 es: ", moda1)

media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("\nLa media de la fabrica 2 es: ", media2)
print("La mediana de la fabrica 2 es: ", mediana2)
print("La moda de la fabrica 2 es: ", moda2)
```

```python

fig,(ax1, ax2) = plt.subplots(nrows =1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
