##1. Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central

```python
# Cargar el archivo csv
autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

```python
# Mostramos los primeros 5 registros del archivo
autos.head()
```

```python
autos.info()
```

```python
# Mostrar valor maximo y minimo de precios
print("Precio Maximo")
autos.price.max()
```

```python
print("Precio Minimo")
autos.price.min()
```

```python
#Determinar el valor de la moda, media y mediana de precios
moda = mode(autos.price)
media = autos['price'].mean()
mediana = autos['price'].median()
print("Valores de precios: \n", "\nModa =", moda, "\nMedia =",media, "\nMediana =", mediana)
```

```python
#Determinar el valor de la desviación estándar de precios
desviacionEstandar = autos['price'].std()
print(f"Desviación estándar: {desviacionEstandar}")
```

```python
# Determinar los estadísticos y cuartiles con la función describe()
autos.price.describe()
```

##2. Realizar un Histograma de precios de los automóviles

```python
# Determinar mediante la regla de sturges la cantidad optima de bins o separadores
import math
n = len(autos.price)
k = int( 1 + math.log2(n))
print("Valor de n = ", n , " \nNumero de intervalos =", k)
```

```python
# Poner títulos y encabezados apropiados
plt.pyplot.hist(autos.price, bins=k)
plt.pyplot.title('Histograma de Precios basado en la regla de Sturges')
plt.pyplot.xlabel('Precios')
plt.pyplot.ylabel('Frecuencia')
plt.pyplot.show()
```

##Interprete los datos y proporcione sus conclusiones
Se puede ver que la mayor frecuencia de la cantidad de precios va desde 5000 a 20000 y va descendiendo la frecuencia, tambien se puede ver que precios entre 25000 y 30000 aparecen con menor frecuencia
